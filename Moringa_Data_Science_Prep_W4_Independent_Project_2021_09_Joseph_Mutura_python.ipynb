# 2.0 Cleaning data

#2.1 Validity

```python
#Lets first get the information of our Data

df.info()
```

```python
#Lets find the shape of our dataset

df.shape
```

Our dataset contains 5000 entries and 25 columns

```python
#Lets drop Scheduled at and Displayed comment culumns as we are not going to use them

df.drop(['Displayed comment', 'Scheduled at'], axis=1, inplace= True)
```

#2.2 Accuracy and Uniformity

```python
#Lets put date in a better format
df['Date'] = df["year"].map(str) +':'+ df["month"].map(str) +':'+ df["day"].map(str)
df.head(4)
```

```python
#Lets put Time in a better format
df['Time'] = df["hour"].map(str) +':'+ df["minute"].map(str) 
df.head(4)
```

```python
#Now we drop all the other redundant columns from the joined date and time

df.drop(['year', 'month','day' ,'hour','minute' ], axis=1, inplace= True)
```

```python
df.head(3)
```

```python
#The data seems accurate and uniform to work with
```

#4.0 Answering the questions

Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

```python
df2 =  df.groupby(['City','Date','Time']).sum()
df2
```

```python
row1 = df2.loc['Paris']
row1
```

```python
row1.sort_values(by=['Bluecar counter'], inplace=True, ascending=False)
```

```python
row1.max()
```
